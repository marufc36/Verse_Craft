```python
import json
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("final_poems_dataset.csv")
```

```python
df.head()
```

```python
poem=df['Poem Content'].tolist()
```

```python
df.isnull().sum()
```

```python
print(type(poem),len(poem))
```

```python
for i in range(0,1):
  print(poem[i])
```

```python
train_df, test_df = train_test_split(poem, test_size=0.1)


```

```python
for i in range(5):
  print(train_df[i])
```

```python
print(len(train_df),len(test_df))
```

```python
for i in range(2):
  print(test_df[i])
```

```python
from transformers import GPT2TokenizerFast
```

```python
tokenizer=GPT2TokenizerFast.from_pretrained("gpt2")
```

```python
tokenizer.pad_token = tokenizer.eos_token
context_length=128
```

```python
training_data=tokenizer(train_df,
                        padding=True,
                        truncation=True,
                        max_length=context_length,
                        return_overflowing_tokens=True,
                        return_length=True,
                        return_tensors="pt")
```

```python
print(training_data["input_ids"][6])
```

```python
Validation_data=tokenizer(test_df,
                        padding=True,
                        truncation=True,
                        max_length=context_length,
                        return_overflowing_tokens=True,
                        return_length=True,
                        return_tensors="pt")
```

```python
print(Validation_data["input_ids"][6])
```

```python
import datasets
```

```python
final_dataset=datasets.DatasetDict({"train": training_data, "valid": Validation_data})
```

```python
type(final_dataset)
```

```python
print(len(final_dataset['train']['input_ids']),len(final_dataset['valid']['input_ids']))
```

```python
from transformers import GPT2LMHeadModel, AutoConfig
```

```python
config= AutoConfig.from_pretrained(
             "gpt2",
             vocab_size=len(tokenizer),
             n_ctx= context_length,
             bos_token_id=tokenizer.bos_token_id,
             eos_token_id=tokenizer.eos_token_id)
```

```python
model=GPT2LMHeadModel(config)
```

```python
from transformers import DataCollatorForLanguageModeling
```

```python
data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False)
```

```python
from transformers import Trainer, TrainingArguments
```

```python
args = TrainingArguments(
        output_dir='output1',
        per_device_train_batch_size=12,
        per_device_eval_batch_size=12,
        evaluation_strategy="steps",
        eval_steps=5_000,
        logging_steps=5_000,
        gradient_accumulation_steps=8,
        num_train_epochs=2,
        weight_decay=0.1,
        warmup_steps=1_000,
        lr_scheduler_type="cosine",
        learning_rate=5e-4,
        fp16=True,
        push_to_hub=False)
```

```python
trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=args,
    data_collator=data_collator,
    train_dataset=final_dataset['train']['input_ids'],
    eval_dataset=final_dataset['valid']['input_ids'])
```

```python
trainer.evaluate()
```

```python
trainer.train()
```

```python
trainer.save_model("poem_generator2")
```

```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast, DataCollatorForLanguageModeling, Trainer, TrainingArguments

# Load the saved model
model_path = "poem_generator3"
model = GPT2LMHeadModel.from_pretrained(model_path)

# Load the tokenizer
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

# Load your training data and prepare it for training (assuming you have it stored in 'final_dataset')

# Define your TrainingArguments
args = TrainingArguments(
    output_dir='output2',  # Change this directory if you want to save the new model elsewhere
    per_device_train_batch_size=12,
    per_device_eval_batch_size=12,
    evaluation_strategy="steps",
    eval_steps=5_000,
    logging_steps=5_000,
    gradient_accumulation_steps=8,
    num_train_epochs=2,  # You can adjust this if needed
    weight_decay=0.1,
    warmup_steps=1_000,
    lr_scheduler_type="cosine",
    learning_rate=5e-4,
    fp16=True,
    push_to_hub=False
)

# Define the DataCollator
data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)

# Define the Trainer
trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=args,
    data_collator=data_collator,
    train_dataset=final_dataset['train']['input_ids'],  # Assuming you have 'final_dataset' prepared
    eval_dataset=final_dataset['valid']['input_ids']   # Assuming you have 'final_dataset' prepared
)

# Continue training
trainer.train()
```
